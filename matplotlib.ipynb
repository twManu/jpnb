# X,Y 刻度, 格線, 儲存

```python
import matplotlib.pyplot as plt
import random
x = range(60)
y1 = [random.uniform(15, 18) for i in x]

plt.figure(figsize=(20, 8), dpi=100)
# plot before X, Y label assignment
plt.plot(x, y1)

# 2.1 adding X, Y axis
x_ticks_labels = ["第{}個".format(i) for i in x]
y_ticks= range(40)
plt.xticks(x[::5], x_ticks_labels[::5])
# number list ok direct add while string needs number list first
plt.yticks(y_ticks[::5])

# 2.2 adding grid
plt.grid(True, linestyle="--", alpha=0.5)

# 2.3 adding label
plt.xlabel("time")
plt.ylabel("temp")
plt.title("temperature over time", fontsize=20)

# 2.4 saving image before show
plt.savefig("save.png")
plt.show()
```

# 繪製多個圖像, 增加圖例

```python
import matplotlib.pyplot as plt
import random
x = range(60)
y1 = [random.uniform(15, 18) for i in x]
y2 = [random.uniform(0, 3) for i in x]

plt.figure(figsize=(20, 8), dpi=100)
# plot before X, Y label assignment
plt.plot(x, y1, label='Y1')
# 2nd plot
plt.plot(x, y2, color='r', linestyle='--', label='Y2')

# 2.1 adding X, Y axis
x_ticks_labels = ["{}th".format(i) for i in x]
y_ticks= range(40)
plt.xticks(x[::5], x_ticks_labels[::5])
# number list ok direct add while string needs number list first
plt.yticks(y_ticks[::5])

# 2.2 adding grid
plt.grid(True, linestyle="--", alpha=0.5)

# 2.3 adding label
plt.xlabel("time")
plt.ylabel("temp")
plt.title("temperature over time", fontsize=20)

# 2.4 adding legend
plt.legend(loc="best")

plt.show()
```

# 多個座標系顯示

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,8), dpi=100)
axes[0].plot(x, y1, label='第一個')
axes[1].plot(x, y2, color='g', label='第二個')
x_ticks_label = ['11點{}分'.format(i) for i in x]
y_ticks = range(40)
axes[0].set_xticks(x[::5])
axes[0].set_yticks(y_ticks[::5])
axes[0].set_xticklabels(x_ticks_label[::5])
axes[1].set_xticks(x[::5])
axes[1].set_yticks(y_ticks[::5])
axes[1].set_xticklabels(x_ticks_label[::5])
```
